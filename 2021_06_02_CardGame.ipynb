```python
import time
import random

Goback = "Yes"
def Opening():
    print("\n\n-------------인생 역전 게임--------------")
    print("당신은 mbs의 유명한 서바이벌 게인인 '해냈어!'에 참가하였습니다.")
    print("드디어 마지막 라운드 도전입니다")
    print("당신 앞에는 천사/악마카드가 뒤집어져 있습니다")
    print("천사카드를 고르면 상금 1,000,000,000을 타게 될 것이고")
    print("악마카드를 고르면 모든 것을 잃고 빈 손으로 돌아갑니다")
def drawAngel():
    print("""
             ||--------||
             ||--^__^--||
             ||--^__^--||
             ||--^__^--||
        ♪♪♪[AngelCard]♪♪♪
        """)
def drawDevil():
    print("""
              ||-------||
              ||--@_@--||
              ||--@_@--||
              ||--@_@--||
              ||-------||
        ※※※[DevilCard]※※※
        """)
def drawPrize():
    print("""  
                 ☆☆☆☆☆☆☆
                 |-------------|
                 |1,000,000,000|
                 |-------------|
                 ☆☆☆☆☆☆☆
                 """)    
def getCardNum():
    cardnum=int(input("\n어떤카드를 고르시겠습니까? (1)or(2) : "))
    if cardnum !=1 and cardnum !=2:
        print("▶▶▶▶▶1,2중에 다시입력하세요◀◀◀◀◀")
        getCardNum()
    else:
        return cardnum
    
def compareCard(cardnum):
    realcard = random.randint(1,2)
    print("당신이 고른 카드는...")
    time.sleep(2)
    
    if cardnum == realcard:
        drawAngel()
        print("축하합니다. 상금 1,000,000,000을 받게되었습니다")
        drawPrize()
        
    if cardnum != realcard:
        drawDevil()
        print("당신은 모든것을 잃고 집으로..")

while Goback != "No":
    
    Opening()
    
    getCardNum()
    
    compareCard(getCardNum)
        
    print()
    
    Goback = input("다시 선택하시겠습니까? Yes or No :")
    while Goback != 'No' and Goback !='Yes':
        print("\n★Yes or No 중에 하나를 골라주세요★")
        Goback = input("다시 선택하시겟습니까? Yes or No : ")      
        
```
